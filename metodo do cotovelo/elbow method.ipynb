```python
import matplotlib.pyplot as plt
from math import sqrt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data-points.csv')
```

```python
sum_of_squares = []
```

```python
for i in range(2, 21):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    sum_of_squares.append(kmeans.inertia_)
```
